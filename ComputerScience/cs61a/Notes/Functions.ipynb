## Call Experssion    

- **Evaluation procedure** for call expressions:
  
  e.g.: `mul(add(1,mul(2,3)),add(4,5))`
  
    1. Evaluate the operator and then operand subexpressions.
    2. Apply the **function** that is the value of the operator subexpression to the **arguments** that are the values of the operand subexpressions.

- Recurrsion in call expressions

## Names, Assignment, and User-Defined Functions

- Assignment:
  - Assignment can be used to give a name to a value.
    - e.g. `area, circ = pi * radius * radius, 2 * pi * radius`
  - Assignment can also be used to give names to functions.
- User-Defined Functions:
  - `def` statement is used to define a function.

## Environment Diagrams

- Executoin rule for assignment statements:
  1. Evaluate the expressions to the right of the `=` operator, from left to right.
  2. Bind all names to the left of the `=` operator to the resulting values in the current frame.


```python
# Example of execution rule:
a = 1
b = 2
b, a = a + b, b
print(b,a)
```

## Defining Funcitons


- Execution rule for `def` statements:
  1. Create a function with signature `<name>(<formal parameters>)`.
  2. Set the body of the function to be everything indented after the `def` statement.
  3. Bind `<name>` to that function in the current frame.

- What is `None`?
  - `None` means return nothing.
  - A function does not explicitly return a value, it implicitly returns `None`. (e.g. `print` function)
  - Python will not directly display `None` when it is the result of an expression.
